```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_Train=pd.read_csv('train_ctrUa4K.csv')
df_Train.head()
```

```python
df_Train.shape
```

```python
df_Test=pd.read_csv('test_lAUu6dG.csv')
df_Test.head()
```

```python
print(df_Train.shape)
print(df_Test.shape)
```

```python
df_Big=df_Train.append(df_Test,sort=False)
df_Big.head()
```

```python
df_Big['Loan_ID']=df_Big['Loan_ID'].str[4:]
```

```python
df_Big.head()
```

```python
df_Big['Loan_ID']=df_Big['Loan_ID'].astype(int)
```

```python
df_Big.dtypes
```

```python
df_Big.isnull().sum()
```

```python
print(df_Big['Gender'].value_counts())
print(df_Big['Married'].value_counts())
print(df_Big['Self_Employed'].value_counts())
```

```python
df_Big['Gender']=df_Big['Gender'].fillna('Male')
df_Big['Married']=df_Big['Married'].fillna('Yes')
df_Big['Self_Employed']=df_Big['Self_Employed'].fillna('No')
```

```python
df_Big.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
df_Big["Gender"]=encoder.fit_transform(df_Big['Gender'])
df_Big["Married"]=encoder.fit_transform(df_Big['Married'])
df_Big["Education"]=encoder.fit_transform(df_Big['Education'])
df_Big["Self_Employed"]=encoder.fit_transform(df_Big['Self_Employed'])
df_Big["Property_Area"]=encoder.fit_transform(df_Big['Property_Area'])
```

```python
df_Big.tail()
```

```python
df_Big.isnull().sum()
```

```python
df_Big['LoanAmount']=df_Big['LoanAmount'].fillna(df_Big['LoanAmount'].mean())
```

```python
print(df_Big['Dependents'].value_counts())
print(df_Big['Loan_Amount_Term'].value_counts())
print(df_Big['Credit_History'].value_counts())
```

```python
df_Big['Dependents']=df_Big['Dependents'].fillna('0')
df_Big['Loan_Amount_Term']=df_Big['Loan_Amount_Term'].fillna('360.0')
df_Big['Credit_History']=df_Big['Credit_History'].fillna('1.0')
```

```python
df_Big.head()
```

```python
df_Big.isnull().sum()
```

```python
df_Big.dtypes
```

```python
df_Big['Loan_Status']=df_Big['Loan_Status'].fillna('Y')
```

```python
df_Big["Loan_Status"]=encoder.fit_transform(df_Big['Loan_Status'])
```

```python
df_Big.head()
```

```python
df_Big['Dependents'].unique()
```

```python
df_Big['Dependents']=df_Big['Dependents'].replace('3+','3')
```

```python
df_Big['Dependents']=df_Big['Dependents'].astype(int)
df_Big['Loan_Amount_Term']=df_Big['Loan_Amount_Term'].astype(float)
df_Big['Credit_History']=df_Big['Credit_History'].astype(float)
df_Big['Loan_Status']=df_Big['Loan_Status'].astype(int)
```

```python
df_Big.dtypes
```

```python
df_Big.head()
```

```python
df_train=df_Big[0:614]
df_test=df_Big[614:]
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_test=df_test.drop(['Loan_Status'],axis=1)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
columns_to_scale = ['Loan_ID', 'ApplicantIncome', 'CoapplicantIncome', 'Loan_Amount_Term', 'LoanAmount']
df_test[columns_to_scale] = scaler.fit_transform(df_test[columns_to_scale])
df_train[columns_to_scale] = scaler.fit_transform(df_train[columns_to_scale])
```

```python
df_train.head()
```

```python
df_train['Loan_Status'].unique()
```

```python
df_test.head()
```

```python
X=df_train.drop(['Loan_Status'],axis=1)
y=df_train[['Loan_Status']]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=8)
```

```python
from catboost import CatBoostClassifier
CB = CatBoostClassifier(iterations=13,
                          learning_rate=1.000001,
                          depth=5)
CB.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
pred=CB.predict(X_test)
accuracy_score(pred,y_test)
```
