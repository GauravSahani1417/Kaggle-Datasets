```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1378, random_state = 3 )
```

```python
from catboost import CatBoostRegressor
CR = CatBoostRegressor(iterations=362,
                          learning_rate=1.0004,
                          depth=7)
CR.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
pred=CR.predict(X_test)

print(r2_score(y_test,pred))
```

```python
prediction=CR.predict(df_Test)
prediction
```
